# Newton's Law of Cooling

```python
# change in temp/ change in time = -K (T obj - T amb )
# K = Newton's constant of proportionality 
```

```python
# Initial conditions
dt = 1 #minutes 
period = 121 #minutes 
T_cup_pred = [68] # degrees Celsius bracket defines a list (not bracket will be an int)
t_cup_pred = range(period) # time in minutes
T_amb = 25.4 # degrees Celsius
K = 0.027
```

```python
# implement the model 
for t in range(1,period): # start from second element bc we already have the first 68
    dT = -K * (T_cup_pred [t-1] - T_amb) * dt
    T_cup_pred_updated = T_cup_pred[t-1] + dT
    T_cup_pred.append(T_cup_pred_updated)
```

```python
t_cup_obs = [0,3,6,9,12,18,24,30,40]
T_cup_obs = [68,64,60.3,56.8,54.5,49.6,46.8,44.1,40.3]
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline 

plt.figure(figsize = (8,5) ) #tupel and size 8 x 5 
plt.plot(t_cup_pred, T_cup_pred, '--k')
plt.plot(t_cup_obs, T_cup_obs, ' ok')
plt.xlabel('Time (minutes)', size=16)
plt.ylabel('Temperature (Celsuis)', size = 16)
plt.show()
```
